# GG4: Neural control with adaptive state estimation

## Overview

In this project, you will interact with a simulated dynamical system inspired by neural interfaces.

The internal state of the system is **not directly observable**. Instead, you can:
- apply **commands** or **stimulation inputs** `u(t)`,
- observe noisy **sensor measurements** `y(t)`,
- and study how the system evolves over time.

From an engineering perspective, this is a problem of **modelling, inference, and control under uncertainty**.

---

## Engineering perspective

The system can be viewed as a **dynamical system**:
- a hidden state evolves over time,
- observations provide partial and noisy information,
- inputs influence the system behaviour.

This leads to three key challenges:
1. **Understanding the system** through exploration,
2. **Extracting useful structure** from observations,
3. **Designing inputs** to influence behaviour.

These ideas are central to many areas of engineering, including control systems, signal processing, inference and machine learning, and emerging neurotechnologies.

## Project structure

Your work is structured in the following stages and you are expected to follow this progression.  
Please do **not** jump straight to control before you have spent time understanding the system behaviour.

- **Week 1 (Group):** Explore the system and build a shared simulation workflow. The Week 1 group code forms the shared foundation for all subsequent individual work.
  **Deliverable:** Group simulation code + brief documentation (**10 marks**).

- **Week 2 (Individual):** Propose and test an estimation approach.  
  **Deliverable:** Individual interim report + supporting code (**20 marks**).

- **Week 3 (Individual):** Design a control strategy and test it in closed loop.  
  *(No formal deliverable; work feeds into the final report.)*

- **Week 4 (Group + Individual):** Compare approaches, test robustness, and prepare presentation.  
  **Deliverables:** Group presentation (**10 marks**) + individual final report (**40 marks**).

## General guidance

This is a **design project**. You are not expected to reproduce a specific method.

Instead, you should:
- explore different approaches,
- justify your design choices,
- evaluate performance carefully,
- and reflect critically on limitations.

Clear reasoning and thoughtful analysis are more important than complexity.

## Project work and expectations

### Supervision and support

- The project leader and/or demonstrator will be available for advice during scheduled sessions (see timetable).

You are expected to maintain a **lab notebook** (physical or electronic), recording:
  - analyses and simulation performed,
  - design decisions,
  - results and observations.
Your notebook may be reviewed during feedback sessions.

You must **keep regular backups** of all code, data, and reports. Loss of work due to computer issues is not considered a valid reason for late submission.

### Use of AI tools
The use of AI-assisted tools is permitted within this project as a support tool for coding assistance, debugging, brainstorming, editing, summarising documentation.

However, students remain fully responsible for understanding all submitted work, validating correctness, interpreting results, and being able to explain and justify all modelling, estimation, and control decisions.

Students must not rely on AI-generated solutions without critical evaluation, or present AI-generated work as independent reasoning.

During demonstrator check-ins and assessments, students will be asked to explain their code, modelling assumptions, experimental choices, and interpretation of results. If a student cannot explain or justify submitted work, this may be treated as poor academic practice.

AI tools should be treated similarly to advanced software tools or online resources: useful for support and productivity, but not a substitute for understanding, engineering judgement, or original analysis.

Please acknowledge how you used AI tools in your report.
